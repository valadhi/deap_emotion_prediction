```python
import os, sys
import pandas as pd
from core.main import *

infodir = 'deap_ledalab_eda'
leda_files_list = os.listdir(infodir)
dict_builder = {}
count = 0
for file in leda_files_list:
    filename, file_extension = os.path.splitext(file)
    filename_pieces = filename.split('_')
    if filename_pieces[-1] == 'era' and file_extension == '.txt':
        participant = filename_pieces[3]
        if int(participant) in exclude_participant:
            continue
        path_to_file = os.path.join(infodir, file)
        df = pd.read_csv(path_to_file, sep='\t')
        extract = df[df.columns[~df.columns.isin(['Event.NID', 'Event.Nr', 'Event.Name'])]]
        extract = extract.iloc[0].to_dict()
        extract['video'] = filename_pieces[1]
        extract['participant'] = participant



        if dict_builder:
            for col, value in extract.items():
                dict_builder[col].append(value)
        else:
            dict_builder = {k: [v] for k,v in extract.items()}

        # print(dict_builder)
        # if count > 2:
        #     import sys
        #     sys.exit(0)
        # count += 1

eda_features = pd.DataFrame.from_dict(dict_builder)
print(eda_features)

print(eda_features.groupby(['participant'])['video'].count())

eda_features.to_pickle("eda_features.pkl")
```
